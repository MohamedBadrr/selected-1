```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
from sklearn.metrics import *


dataset = pd.read_csv('E:\\ProjectSelected\\LogisticReg\\diabetes.csv',header=None , skiprows=1,names=['Pregnancies','Glucose',
                                                                   'BloodPressure','SkinThickness',
                                                                   'Insulin','BMI','DiabetesPedigreeFunction',
                                                                   'Age','result'])
# dataset.head(20)
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 44)


# Feature Scaling
# from sklearn.preprocessing import StandardScaler
# sc = StandardScaler()
# X_train = sc.fit_transform(X_train)
# X_test = sc.transform(X_test)


LogR = LogisticRegression(max_iter=1000)
# LogR = LogisticRegression(max_iter=1000,C=.5)

history = LogR.fit(X_train, y_train)
y_pred = LogR.predict(X_test)
for i in range(0,len(y_test)):
    print('The real Result is = ',np.array((y_test)).T[i] ," =>>", 'The Predicted Result is = ',y_pred[i] )
```

```python
#Confusion Matrix
CM=confusion_matrix(y_pred, y_test)
print('confusion_matrix is : \n',CM)
print('Train accurcy is : ' , LogR.score(X_train, y_train)*100)
print('Test accurcy is : ' , LogR.score(X_test, y_test)*100)
print('Number of iterations is : ' , LogR.n_iter_)
print('----------------------------------------------------')
acc = accuracy_score(y_test, y_pred)
print('accuracy Score is : ', acc*100)
print('................................................')


print("ROC Curve")

fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred)
auc = metrics.roc_auc_score(y_test, y_pred)
#create ROC curve
plt.plot(fpr,tpr,label="AUC="+str(auc))
plt.title("ROC Curve")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc=4)
plt.show()
```
